```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs 
```

```python
data= make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data 
```

```python
data[0]
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow',alpha=5)
```

```python
data[1] #cluster that it belongs to
```

```python
from sklearn.cluster import KMeans # Unsupervised
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ #Predicted labels
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,sharey=True, figsize=(10,6))  # one by two

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
